```python


lower_number = 1042000
higher_number = 702648265

for num in range(lower_number, higher_number + 1):

   
   order = len(str(num))
    
   
   sum = 0

   temp = num
   while temp > 0:
       digit = temp % 10
       sum += digit ** order
       temp //= 10
   
   if num == sum:
       print('first armstrong num is',+ num)
       break
```
