```python
lst=[]
print(lst)


#adding integer into the list

lst=[100,200,2000]
print(lst)

#adding mixed datatypes into the list
lst=[2,'Declathon',12.0]
print(lst)

#list having another list
lst=[lst,[12,20,1444]]
print(lst)

#list indexing- returns the index of the first matched item
print(lst[1])

print(lst[1][2])

#functions of list data structure
lst.append('Action')
print(lst)


#Add all elements of a list to the another list
lst1=['almond', 'olive']
lst.extend(lst1)
print(lst1)

# Insert an item at the defined index
lst.insert(0,123)
print(lst)

#Removes an item from the list
#lst.remove(12)
#print(lst)

#Removes and returns an element at the given index
lst.pop(3)
print(lst)

#Removes all the items from the list
#lst.clear()
#print(lst)Returns the copy of list

#Reverse the items in the index
lst.reverse()
print(lst)

#Returns the copy of list
lst2=lst.copy()
print(lst2)
```

```python
dit={}

print(dit)

#adding value in to dictionary
dit={1: 'football', 2: 'Hockey'}
print(dit)

# dictionary with mixed keys
dit = {'sports': 'cricket', 1: [2, 4, 3]}
print(dit)


# using dict()
dit1 = dict({1:'music', 2:'sports'})

print(dit1)

# from sequence having each item as a pair
dit2 = dict([(1,'apple'), (2,'cake')])
print(dit2)

#accessing values from dictionary
print(dit['sports'])
print(dit1[1])

#removing elements from dictionary
print(dit2.pop(2))
print(dit2)

print(dit1.popitem())

#deleting dictionary
#del dit2


#key method
print(dit.keys())

#value method
print(dit.values())

#update method
dit3={2:'baseball'}
dit.update(dit3)
print(dit)
```

```python
set1={6,12,24}

print(set1)

#adding mixed datatypes into set
set1 = {1.0, "Hello", (1, 2, 3)}
print(set1)


#add method
set1.add(10)
print(set1)

# update method
set1.update([18, 28, 48])
print(set1)

#discard method
set1.discard(48)
print(set1)

#remove method
set1.remove(1.0)
print(set1)

#intializing using set method

set2 = set("Spaceship")
print(set2)
```
